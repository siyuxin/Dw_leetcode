### 7 Fib

```python
class Solution:
    def Fibonacci(self, n):
        # write code here
        if n <= 1:
            return n
        first, second, third = 0, 1, 0
        for i in range(2, n+1):
            third = first + second
            first = second
            second = third
        return third
if __name__ == '__main__':
    s = Solution()
    print(s.Fibonacci(5))
```

```python
class Solution:
    def Fibonacci(self, n):
        # write code here
        res=[0,1,1,2]
        while len(res)<=n:
            res.append(res[-1]+res[-2])
        return res[n]
if __name__ == '__main__':
    s = Solution()
    print(s.Fibonacci(5))
```
